### Exercice 1
### Computing Pi avec Montecarlo

```python
def estimate_pi(N):
    x = [np.random.uniform(-1,1) for i in range(N)]
    y = [np.random.uniform(-1,1) for i in range(N)]
    in_circle = [1 if np.sqrt(x[i]**2 + y[i]**2) <= 1 else 0 for i in range(N)]
    res = (sum(in_circle)/N) * 4
    return (res, x, y)
```

```python
pi = estimate_pi(5000)
print(pi[0])
```

```python
def bonus(x,y):
    c = plt.Circle((0,0),1, fill=False, color='orange')
    fig, ax = plt.subplots()
    ax.set_aspect(1)
    ax.add_artist(c)
    ax.plot([-1,1], [1,1], color='green')
    ax.plot([-1,-1], [-1,1], color='green')
    ax.plot([-1,1], [-1,-1], color='green')
    ax.plot([1,1], [1,-1], color='green')
    plt.xlim(-1.1, 1.1)
    plt.ylim(-1.1, 1.1)
    
    N = len(x)
    x_blue = []
    y_blue = []
    x_red = []
    y_red = []
    for i in range(N):
        if np.sqrt(x[i]**2 + y[i]**2) <= 1 :
            x_blue.append(x[i])
            y_blue.append(y[i])
        else:
            x_red.append(x[i])
            y_red.append(y[i])
    
    ax.scatter(x_blue,y_blue, s=1)
    ax.scatter(x_red,y_red, s=1, color='red')
       
    plt.show()
```

```python
x, y = pi[1], pi[2]
bonus(x,y)
```

### Les questions 1 √† 3 sont √† la fin du pdf
L‚Äôalgorithme d'echantillonnage par transformation inverse consite √† tirer des x al√©atoires entre 0 et 1 qui sont en fait nos p(x). Ensuite, il suffit de trouver √† quelles valeures ces probabilit√©s sont associ√©es pour avoir la distribution

```python
l=3
N=5000
```

```python
def echantillons(N, l):   
    x = np.random.uniform(0,1,5000)
    z = [-l * math.log( 1 - i) for i in x]  
    return z
```

```python
b = echantillons(N,l)
```

```python
plt.hist(b)
plt.show()
```

```python
c = np.random.exponential(l,N)
```

```python
N = 10000
l = 3
def t(x):
    return x**2
```

```python
def q7(N,l):
    x = np.random.exponential(l,N)
    y = [t(i) for i in x]
    m = np.mean(y)
    err = np.sqrt(sum( [((i-m)**2) for i in y] )/N )
    return m,err
```

```python
l_res = [q7(N,l) for i in range(N)]
```

```python
dqidoj = np.asarray(l_res)
print(dqidoj[:,1])
```
